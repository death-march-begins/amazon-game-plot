```python
#import
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
#read file
file = open("game-review.json").read()
list_data = file.split("\n")
```

```python
list_data
```

```python
#declar
score=[]
produk=[]
```

```python
top = json.loads(list_data[0])
```

```python
top
```

```python
for x in range(0, 20) :
    data = json.loads(list_data[x])
    temp_score = data['score'] / top['score']
    score.append(float(temp_score))
    produk.append(data['produk'])
```

```python
#create bar
plt.bar(produk, score, color='orange')

# Add title and axis names
plt.title("Top 20 Amazon Gaming Product Reviews 2018")
plt.xlabel('Product')
plt.ylabel('Score')

#insert word and rotate word
plt.xticks( produk, rotation=45, horizontalalignment='right')

#set margin bottom
plt.subplots_adjust(bottom= 0.2)
```
